```python
from __future__ import print_function

import codecs
import os
import gensim #워드를 벡터형태로 바꿔주는 라이브러리 
```

```python
%%time
def read_data(filename):
    with codecs.open(filename, 'r', 'utf-8') as f:
        data = [line.split(' ') for line in f.read().splitlines()]
    return data
        
data = read_data('/content/gdrive/My Drive/Colab Notebooks/wiki/wiki.txt')

print(len(data))
print(data[0])
```

```python
%%time
import gensim
model = gensim.models.Word2Vec(data, min_count=1, size=100, window=5, sg=1) #split한 데이터를 이용 size는 생성해서 만들어질 차원 sg=1 스킵그램 0이면 cbow 
model.save('/content/gdrive/My Drive/Colab Notebooks/wiki/wiki.w2v')
```

```python
# model = gensim.models.Word2Vec.load(os.path.join('/content/gdrive/My Drive/Colab Notebooks/nlp/wiki', 'wiki_notoken.w2v'))
model = gensim.models.Word2Vec.load(os.path.join('/content/gdrive/My Drive/Colab Notebooks/wiki', 'wiki.w2v'))
```

```python
model.wv[u'고양이'] #차원 100 
```

```python
model.wv.similarity(u'고양이', u'개')
```

```python
for t in model.wv.most_similar(u'강아지'):
    print(t[0])
```

```python
result = model.wv.accuracy(os.path.join('/content/gdrive/My Drive/Colab Notebooks/wiki','word_analogy_korean.txt')) 
for r in result:
    print('%s\t%s\t%s' % (r['section'], len(r['correct']), len(r['incorrect'])))
```
